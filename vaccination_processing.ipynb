```python
import pandas as pd
```

```python
df_v = pd.read_csv("country_vaccinations.csv")
df_v_C= df_v[(df_v['country'] == 'China')].iloc[:, :6]
df_v_US = df_v.drop(df_v_C.index, axis =0).fillna(0)
#df_v_US = df_v_US.drop(columns=['iso_code','total_vaccinations','people_vaccinated']).fillna(0)
df_v_US = df_v_US[["country","date","people_fully_vaccinated"]]
df_v_US.rename(columns= {'country':'Country/Region','date':'ObservationDate','people_fully_vaccinated':'People_Fully_Vaccinated'},inplace=True)
df_v_US['ObservationDate'] = pd.to_datetime(df_v_US['ObservationDate'],dayfirst=True,format='%Y-%m-%d')
df_v_US = df_v_US.groupby(['ObservationDate']).sum().reset_index()
df_v_US.sort_values('ObservationDate')
```

```python
df_v_US['Dates'] = df_v_US.index
df_v_US=df_v_US.drop(columns=['ObservationDate'])
df_v_US
```

```python
cols = df_v_US.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_v_US = df_v_US[cols]
df_v_US
```

```python
df_v_US.to_csv("Vaccination_US.csv")
```
