```python
import pandas as pd
```

```python
df_c = pd.read_csv('../Data/covid_us_ext.csv')
df_c_C= df_c[(df_c['Country/Region'] == 'Mainland China')].iloc[:, :6]
#df_c_US = df_c.drop(df_c_C.index, axis =0)
df_c_US = df_c[(df_c['Country/Region'] == 'US')]
df_c_US['ObservationDate'] = pd.to_datetime(df_c_US['ObservationDate'],dayfirst=True,format='%m/%d/%Y')
df_c_US.sort_values('ObservationDate')
```

```python
df_c_US = df_c_US.groupby(['ObservationDate']).sum().reset_index()#.sort_values('ObservationDate')
df_c_US = df_c_US.drop(columns=['SNo','Deaths','Recovered'])

df_c_US
```

```python
df_c_US.to_csv("../Data/Confirmed_Cases_US.csv")
```
