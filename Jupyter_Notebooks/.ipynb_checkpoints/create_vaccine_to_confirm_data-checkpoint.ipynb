```python
import pandas as pd

df_v = pd.read_csv("country_vaccinations.csv")
df_v_C= df_v[(df_v['country'] == 'China')].iloc[:, :6]
df_v_US = df_v.drop(df_v_C.index, axis =0).fillna(0)
#df_v_US = df_v_US.drop(columns=['iso_code','total_vaccinations','people_vaccinated']).fillna(0)
df_v_US = df_v_US[["country","date","people_fully_vaccinated"]]
df_v_US.rename(columns= {'country':'Country/Region','date':'ObservationDate','people_fully_vaccinated':'People_Fully_Vaccinated'},inplace=True)
df_v_US['ObservationDate'] = pd.to_datetime(df_v_US['ObservationDate'],dayfirst=True,format='%Y-%m-%d')
df_v_US = df_v_US.groupby(['ObservationDate']).sum().reset_index()
df_v_US.sort_values('ObservationDate')
```

```python
df_c = pd.read_csv('covid_19_data.csv')
df_c_C= df_c[(df_c['Country/Region'] == 'Mainland China')].iloc[:, :6]
df_c_US = df_c.drop(df_c_C.index, axis =0)
#df_c_US = df_c[(df_c['Country/Region'] == 'US')]
df_c_US['ObservationDate'] = pd.to_datetime(df_c_US['ObservationDate'],dayfirst=True,format='%m/%d/%Y')
df_c_US.sort_values('ObservationDate')
```

```python
df_c_US = df_c_US.groupby(['ObservationDate']).sum().reset_index()#.sort_values('ObservationDate')
df_c_US = df_c_US.drop(columns=['SNo','Deaths','Recovered'])
df_c_US
```

```python
df_US = pd.merge(df_c_US, df_v_US, on='ObservationDate', how='outer')
df_US
```

```python
df_as_test = df_US.loc[(df_US['ObservationDate'] > '2021-02-27')].reset_index().drop(columns=['index'])
df_as_test['People_Fully_Vaccinated'].to_csv('02_28_to_03_20_vaccination.csv')
```

```python
df_US=df_US.loc[(df_US['ObservationDate']>= '2020-12-20') & (df_US['ObservationDate'] <= '2021-02-27')]
df_v_US[df_v_US['ObservationDate']=='2021-02-15'] = 14077440
df_US = df_US.reset_index().drop(columns=['index'])
df_US
```
