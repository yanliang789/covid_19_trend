```python
import pandas as pd
df_v = pd.read_csv("../Data/vaccination_us_ext.csv")
df_v
```

```python
df_v_US = df_v[['date','location','people_fully_vaccinated']].copy()
df_v_US
```

```python

df_v_US = df_v_US.fillna(0)
df_v_US['date'] = pd.to_datetime(df_v_US['date'],dayfirst=True,format='%Y-%m-%d')
df_v_US = df_v_US.groupby(['date']).sum().reset_index()
df_v_US.columns  =['ObservationDate','People_Fully_Vaccinated']
df_v_US.sort_values('ObservationDate')
```

```python
df_v_US.to_csv("../Data/Vaccination_US.csv")
```
